```python
p=df['Urban Pop %']
p2=list(p)
p2
```

```python
k=[]
for i in p2:
    r=""
    for j in i:
        if(j.isdigit()):
            r=r+j
    k.append(r)
k
```

```python
import numpy as np
```

```python
df['Urban Pop %']=k
df
```

```python
df['Urban Pop %']=df['Urban Pop %'].replace('',0)
df
```

```python
df['Urban Pop %']=df['Urban Pop %'].astype('int')
df
```

```python
t=100-df['Urban Pop %']
t2=list(t)
t2
```

```python
df['Rpopulation']=t2
```
